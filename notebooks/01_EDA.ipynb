```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your combined dataset from the data folder
file_path = '../data/combined_standings_2024.csv'
df = pd.read_csv(file_path)

# Look at the first 5 rows
df.head()
```

```python
df.info()
```

```python
# Create a scatter plot to visualize the relationship between xG and Points
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='xG', y='Pts', hue='League')
plt.title('Expected Goals (xG) vs. Final Points')
plt.xlabel('Total Expected Goals (xG) for the Season')
plt.ylabel('Total Points (Pts)')
plt.grid(True)
plt.show()
```
